Instanciar modelo BERTweet de Huggingface para el análisis de sentimientos. El modelo devuelve una etiqueta con la clasificación del sentimiento ('NEU', 'NEG', 'POS') y el índice o puntuación de confianza asociado

```python
sentiment_analyzer = pipeline("sentiment-analysis", model="finiteautomata/bertweet-base-sentiment-analysis")
```

# **CONJUNTO APPLE**

Primero se carga el dataset inicial con los tweets categorizados por tópico
 y se transformarla columna 'body' en tipo string para la entrada al modelo

```python
 data_apple = pd.read_csv('/content/drive/MyDrive/TFG ADE/AppleData15Topics.csv')
text_apple = (data_apple['body'].astype(str))
```

Se ejecuta el modelo y se añaden dos columnas para guardar en el dataset la categorización del sentimiento y la puntuación correspondiente

```python
# APPLE
data_apple ['Sentimiento'] =text_apple.apply(lambda x: sentiment_analyzer(x)[0]['label'])
data_apple ['Puntuación'] = text_apple.apply(lambda x: sentiment_analyzer(x)[0]['score'])
```

Se observa la categorización resultante

```python
data_apple.tail()
```

Se guarda el dataset para no perderlo cuando expire la sesión

```python
data_apple.to_csv('/content/drive/MyDrive/TFG ADE/AppleDataSentimientos.csv')
```
